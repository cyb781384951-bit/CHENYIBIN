```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
```

```python
mnist = keras.datasets.mnist
(train_images,train_labels),(test_images,test_labels) =\
mnist.load_data()
```

```python
print('train_images.shape =', train_images.shape)
print('test_images.shape =', test_images.shape)
print('train_labels.shape =', train_labels.shape)
print('test_labels.shape =', test_labels.shape)
```

```python
num = train_images[6000]
for i in range(28):
    for j in range(28):
      print('{:4d}'.format(num[i][j]), end='')
    print()
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(4, 1, 1)
ax2 = fig.add_subplot(4, 1, 2)
ax3 = fig.add_subplot(4, 1, 3)
ax4 = fig.add_subplot(4, 1, 4)
ax1.imshow(train_images[1000], cmap='Greys') # 显示第一张训练图像
ax2.imshow(train_images[2000], cmap='Greys') # 显示第二张训练图像
ax3.imshow(train_images[3000], cmap='Greys') # 显示第三张训练图像
ax4.imshow(train_images[4000], cmap='Greys') # 显示第四张训练图像
```

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28, 28)),
    keras.layers.Dense(128, activation='relu'),
    keras.layers.Dense(10, activation='softmax')
])
```

```python
model.summary()
```

```python
model.compile(optimizer='sgd',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy'])
train_images, test_images = train_images / 255, test_images / 255
model.fit(train_images, train_labels, epochs=10, verbose=1)

test_loss, test_acc = model.evaluate(test_images, test_labels,
    verbose=0)
```

```python
print('\n测试准确度:', test_acc)
```

```python
model.compile(optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy'])
train_images, test_images = train_images / 255, test_images / 255
model.fit(train_images, train_labels, epochs=2, verbose=1)

test_loss, test_acc = model.evaluate(test_images, test_labels,
    verbose=0)
```

```python
print('\n测试准确度:', test_acc)
```
